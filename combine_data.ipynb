```python
import pandas as pd
import numpy as np
import re
```

```python
df_champion = pd.read_csv("data/df_champion_clean.csv")
df_champion.head()
df_champion.tail()
```

```python
F1_driver = pd.read_csv("data/F1Drivers_Dataset.csv")
F1_driver.head()
```

```python
select_champion_F1 = F1_driver[F1_driver["Champion"]]
selected_row = ["Driver", "Championship Years"]
select_champion_F1[selected_row]
```

```python
select_champion_F1['Championship Years'] = select_champion_F1['Championship Years'].apply(lambda x: re.sub(r"[\[\]]", "", x))
select_champion_F1['Championship Years'] = select_champion_F1['Championship Years'].apply(lambda x: x.split(","))
#select_champion_F1['Championship Years'].apply(lambda x: x.replace("[" or "]", ""))
```

```python
select_champion_F1 = select_champion_F1.explode("Championship Years")
```

```python
select_champion_F1['Championship Years'] = select_champion_F1['Championship Years'].apply(lambda x: int(x))
```

```python
merged_dataset_F1 = pd.merge(df_champion, select_champion_F1, right_on="Championship Years", left_on="Year")
```

```python
merged_dataset_F1.columns.tolist()
```

```python
merged_dataset_F1 = merged_dataset_F1.drop(columns=['Championship Years', 'Champion', 'Driver'])
```

```python
merged_dataset_F1.head()
```

```python
merged_dataset_F1.to_csv("merged_dataset_F1.csv")
```
